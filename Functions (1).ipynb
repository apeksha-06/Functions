7. What are the  advantages of using geneartors over regular functions?




-> Advantages of generators over regular function:
- Memory efficiency:-
Generators produce one item at atime instead of storing all results in memory, making them ideal for large datasets or infinite sequences.
- Lazy evaluation:- Generators compute values only when needed. regular functions compute everything immediately , even if you don't use all results.
-Infinite sequences:- regular function cannot return infinite data(they'd never finish).
generators can yield indefinitely, making them perfect for infinite streams(like sensor data, log processing,etc.)
- Faster start time:- generators start producing output immediately without waiting for all calcualations to finish.

```python
#4 Write a python function that checks if a given number is prime or not from 1 to 200.
from math import sqrt

def primeNo(num):
  if num <= 1:
    print(f"{num} is not a prime number")
    return False  # Numbers less than or equal to 1 are not prime
  for i in range(2, int(sqrt(num)) + 1):
    if num % i == 0:
      print(f"{num} is not a prime number")
      return False  # Found a divisor, so not prime
  print(f"{num} is a prime number")
  return True # No divisors found, so it's prime

num = int(input("enter a number between (1 to 200):"))
check = primeNo(num)
```

```python
#5 Create an iterator class in python that generates the fibonacci sequence up to a specified number of terms.
def fibonacci_generator(n_terms):

    a, b = 0, 1  # Initialize the first two Fibonacci numbers
    count = 0

    while count < n_terms:
        yield a  # Yield the current Fibonacci number
        a, b = b, a + b  # Update to the next pair of Fibonacci numbers
        count += 1

fib_gen = fibonacci_generator(10)

print("First 10 Fibonacci numbers:")
for num in fib_gen:
    print(num, end=" ")
```

```python
#6 Write a generator function in python that yields the power of 2 up to a given exxponent.
def max_exponent(num):
    for i in range(0, num + 1):
        a = 2 ** i
        yield a

num = int(input("enter the power of exponent"))
gen = max_exponent(num)

for _ in range(num + 1):
    print(next(gen))
```

```python
#7 Implement a generator function that reads a file line by line and yields each line as a strings.
def read_lines(file_content):
    for line in file_content:
        yield line.strip()  # remove any extra spaces/newlines

# Simulated file content as a list of strings
file = [
    "This is the first line\n",
    "This is the second line\n",
    "This is the third line\n"
]

# Using the generator
for line in read_lines(file):
    print(line)
```

```python
#9 Write a python program that uses "map()" to convert a list of temperatures from celsius to fahrenheit.
celsius =[0,20,38,100]
# using map() with lambda
fahrenheit = list(map(lambda x: (x * 9/5) + 32, celsius))
print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
```

```python
#10 Create a python program that uses "filter()" to remove all the vowels from a  given string.
string= input("enter the string:")
new_string=string.lower()
vowels = ['a', 'e','i','o','u']
filtered_string = ''.join(filter(lambda x: x not in vowels, new_string))

print("Original string:", string)
print("Filtered string:", filtered_string)
```
